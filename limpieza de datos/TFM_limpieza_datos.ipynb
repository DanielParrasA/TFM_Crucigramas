```python
import pandas as pd
```

```python
dataframes = []

for i in range(1, 105):  
    filename = f"a{i}_TFM.csv"  
    df = pd.read_csv(filename)  
    dataframes.append(df)  

merged_df = pd.concat(dataframes, ignore_index=True)
df = merged_df
df
```

```python
def limpiar_frecuencia(frecuencia):
    frecuencia = frecuencia.lower()  # Convertir a minúsculas para ser insensible a mayúsculas
    if "bajo" in frecuencia or "baja" in frecuencia:
        return "Bajo"
    elif "alto" in frecuencia or "alta" in frecuencia:
        return "Alto"
    elif "medio" in frecuencia or "media" in frecuencia:
        return "Medio"
    else:
        return None  # Si no es Alto, Bajo o Medio
```

```python
df["Frecuencia"] = df["Frecuencia"].apply(limpiar_frecuencia)

cantidad_none = df["Frecuencia"].isna().sum()
cantidad_none
```

```python
registros_con_none = df[df["Frecuencia"].isna()]
registros_con_none
```

```python
df["Frecuencia"].fillna("Bajo", inplace=True)
cantidad_none = df["Frecuencia"].isna().sum()
cantidad_none
```
