```python
df_junto['Valida'] = True

df_junto.loc[
    df_junto['Temperatura 0.2'].str.startswith('Con una temperatura de') |
    df_junto['Temperatura 0.7'].str.startswith('Con una temperatura de'),
    'Valida'
] = False
```

```python
df_junto['Palabra'] = df_junto['Palabra'].str.replace('ñ', 'n')
df_junto['Palabra'] = df_junto['Palabra'].str.replace('á', 'a')
df_junto['Palabra'] = df_junto['Palabra'].str.replace('é', 'e')
df_junto['Palabra'] = df_junto['Palabra'].str.replace('í', 'i')
df_junto['Palabra'] = df_junto['Palabra'].str.replace('ó', 'o')
df_junto['Palabra'] = df_junto['Palabra'].str.replace('ú', 'u')
df_junto['Palabra'] = df_junto['Palabra'].str.replace('ü', 'u')
df_junto['Palabra'] = df_junto['Palabra'].str.replace('-', '')
```

```python
palabras_con_especiales = df_junto[df_junto['Palabra'].str.contains(r"[!@#$%^&*(),.?':{}|<>]-", regex=True)]

palabras_con_especiales
```

```python
df_junto['Letras'] = df_junto['Palabra'].apply(lambda palabra: len(palabra))
```
